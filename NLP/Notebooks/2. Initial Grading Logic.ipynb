```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

model = SentenceTransformer('all-MiniLM-L6-v2')

def grade_answer(student_answer, key_answer, max_marks=10):
    # Encode both answers
    embeddings = model.encode([student_answer, key_answer])
    
    # Compute cosine similarity
    sim = cosine_similarity([embeddings[0]], [embeddings[1]])[0][0]
    
    # Convert similarity to marks
    if sim >= 0.9:
        marks = max_marks
    elif sim >= 0.75:
        marks = round(max_marks * 0.8)
    elif sim >= 0.6:
        marks = round(max_marks * 0.5)
    else:
        marks = 0
    
    return marks, sim
```

```python
key_answer = "Sadiq mansoor is son of Qasim Mansoor."
student_answer = "Plants make food using sunlight and carbon dioxide"
marks, sim = grade_answer(student_answer, key_answer)
print(f"Similarity Score: {sim:.2f}")
print(f"Marks Awarded: {marks}/10")
```
