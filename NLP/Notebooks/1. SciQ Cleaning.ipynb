```python
import pandas as pd

# Load all 3 files
train = pd.read_csv("D:\Career\FYP\Dataset/train.csv")
valid = pd.read_csv("D:\Career\FYP\Dataset/validation.csv")
test  = pd.read_csv("D:\Career\FYP\Dataset/test.csv")

# Merge into one dataframe
df = pd.concat([train, valid, test], ignore_index=True)

# Keep only question + correct_answer
df = df[['question', 'correct_answer']]

# Remove duplicates (if any)
df = df.drop_duplicates()

# Reset index
df = df.reset_index(drop=True)

# Save as a clean CSV
df.to_csv("D:\Career\FYP\Dataset/SciQ.csv", index=False)

print("✅ Cleaned dataset saved as data/science_keys.csv")
print(df.head())
```
